**Visualisasi Data**

```python
def plot(col, df=df):
    # Membuat plot batang
    ax = df.groupby(col)['LUNG_CANCER'].value_counts(normalize=True).unstack().plot(kind='bar', figsize=(8,5))

    # Anotasi setiap batang dengan nilainya
    for p in ax.patches:
        ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

    return ax

def set_legend_and_ticks(feature_name):
    # Memplot distribusi fitur yang ditentukan
    ax = plot(feature_name)

    # Mendapatkan label tick saat ini
    ticks = ax.get_xticks()

    # Menetapkan label tick baru
    new_ticks = ['NO '+feature_name, feature_name]
    ax.set_xticklabels(new_ticks)

    # Mendapatkan objek legenda
    legend = ax.legend()

    # Menetapkan label baru untuk legenda
    legend_labels = ['No', 'Yes']
    
    # Menetapkan judul legenda
    legend.set_title('Have lung cancer')
    
    # Menetapkan label baru untuk legenda
    for i, label in enumerate(legend_labels):
        
        # Mengubah label legenda
        legend.get_texts()[i].set_text(label)

    # Menampilkan plot
    plt.show()
```

```python
# Menghapus kolom yang tidak diperlukan
df_new=df.drop(columns=['GENDER','AGE', 'SMOKING', 'SHORTNESS OF BREATH'])

# Menampilkan data yang sudah dihapus kolomnya
df_new
```

Mencari korelasi

```python
# Membuat variabel baru untuk korelasi, menggunakan dataframe baru
cn=df_new.corr()
# Menampilkan tabel korelasi
cn
```

```python
# Membuat visualisasi korelasi antar kolom dalam bentuk heatmap
cmap=sns.diverging_palette(260,-10,s=50, l=75, n=6, as_cmap=True) # palet warna
plt.subplots(figsize=(18,18)) # figsize untuk menentukan ukuran gambar
sns.heatmap(cn,cmap=cmap,annot=True, square=True)

# Menampilkan visualisasi korelasi antar kolom dalam bentuk heatmap
plt.show()
```

Dari matriks korelasi, terlihat bahwa ANXIETY dan YELLOW_FINGERS berkorelasi lebih dari 0.5. Jadi, keduanya akan digabungkan dan dinamakan 'ANXYELFIN'

```python
# Menggabungkan kolom 'YELLOW_FINGERS' dan 'ANXIETY' menjadi satu kolom baru 'ANXYELFIN'
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
# Memisahkan variabel independen dan dependen
X = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

Target Distribution Imbalance Handling

```python
from imblearn.over_sampling import ADASYN # menambahkan library imblearn (imbalance)

# Membuat objek ADASYN
adasyn = ADASYN(random_state=42)

# Melakukan resampling data
X, y = adasyn.fit_resample(X, y)

# Menampilkan jumlah data setelah resampling
len(X)
```

**Pemilihan Model**

Logistic Regression

```python
# Memisahkan data menjadi data training dan data testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
# Mengaplikasikan model Logistic Regression ke dalam data training
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)

# Melakukan training model
lr_model.fit(X_train, y_train)
```

**Evaluasi Model**

```python
# Memprediksi hasil menggunakan data testing
y_lr_pred= lr_model.predict(X_test)

# Menampilkan hasil prediksi
y_lr_pred
```

```python
# Menghitung akurasi model
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)

# Menampilkan hasil akurasi model
print(lr_cr)
```

Model ini **98%** akurat.

1.	Akurasi model mendapatkan nilai 98%. Hal ini menunjukkan kemampuan model dalam memprediksi dengan tepat 98% dari total data yang diamati. Dengan begitu, dapat dikatakan bahwa model memiliki kinerja yang sangat baik dalam mengklasifikasikan sentimen.
2.	Presisi untuk kelas sentimen negatif dan positif (0 dan 1) adalah 98%. Hal ini menunjukkan bahwa sebagian besar prediksi yang diklasifikasikan sebagai sentimen positif atau negatif oleh model adalah benar.
3.	Recall untuk kedua kelas, baik sentimen positif maupun negatif bernilai 98%. Hal ini juga menunjukkan bahwa model dapat mengidentifikasi sebagian besar sampel yang benar-benar termasuk dalam setiap kelas dengan benar.
4.	F1-score untuk kedua kelas (0 dan 1) adalah 98%. F1-score adalah rata-rata harmonis dari presisi dan recall yang dapat memberi gambaran  performance model dalam mengklasifikasikan sentimen.
